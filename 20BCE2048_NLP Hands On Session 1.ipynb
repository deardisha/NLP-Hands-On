# Finding Frequency Distribution of Lincoln's 1861 Inaugural Speech with words having length > 3 in an alphabetical order

```python
t1 = inaugural.words('1861-Lincoln.txt')
```

```python
data_analysis = nltk.FreqDist(t1)
```

```python
# Let's take the specific words only if their length is greater than 4.
filter_words = dict([(m, n) for m, n in data_analysis.items() if len(m) > 4])
```

```python
for key in sorted(filter_words):
    print("%s: %s" % (key, filter_words[key]))
```

```python
data_analysis = nltk.FreqDist(filter_words)
```

```python
 data_analysis.plot(25, cumulative=False)
```

# Creating a word cloud for a paragraph from Lincoln's 1861 Inaugural Speech 

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 

text = 'Fellow-Citizens of the United States: In compliance with a custom as old as the Government itself, I appear before you to address you briefly and to take in your presence the oath prescribed by the Constitution of the United States to be taken by the President before he enters on the execution of this office." I do not consider it necessary at present for me to discuss those matters of administration about which there is no special anxiety or excitement.'

def generate_wordcloud(text): # optionally add: stopwords=STOPWORDS and change the arg below
    wordcloud = WordCloud(
                          width=800, height=400,
                          relative_scaling = 1.0,
                          stopwords = {'to', 'of', 'the', 'you', 'by', 'i', 'a', 'as'} # set or space-separated string
                          ).generate(text)
    
    fig = plt.figure(1, figsize=(8, 4))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()

generate_wordcloud(text)
```
